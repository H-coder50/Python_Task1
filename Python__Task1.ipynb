# Problem 1: Calculate Distance Using Speed and Time
Problem Statement: Write a program to calculate the distance traveled by an object, given
its speed and the time it travels.

Formula:
Distance = Speed × Time

Input: Speed (in meters per second) and time (in seconds).

Output: Distance traveled (in meters).

Example:

Input: Speed = 10, Time = 5

Output: Distance = 50 meters

```python
speed = int(input("Enter speed (m/s): "))
time = int(input("Enter time (s): "))
Distance = speed * time
print(f"Distance traveled by an object: {Distance} m")
```

```python
# Define a function to calculate distance
def calculate_distance(speed, time):
    return speed * time

# Input values
speed = 10
time = 5

# Call the function
distance = calculate_distance(speed, time)

# Output the result
print(f"Distance traveled by an object: {distance} m")
```

Force = Mass × Acceleration

Output: Force (in Newtons).

Example:

Input: Mass = 5, Acceleration = 3

# Problem 3: Gravitational Potential Energy

Problem Statement: Write a program to calculate the gravitational potential energy of an
object.

Formula:
Potential Energy = m × g × h

where g = 9.8 m/s² (acceleration due to gravity), m is the mass, and h is the height.

Input: Mass (in kilograms) and height (in meters).

Output: Gravitational potential energy (in joules).

Example:

Input: Mass = 2, Height = 10

Output: Potential Energy = 196 J

```python
Mass = int(input("Enter the Mass of an object: kg"))
Height = int(input("Enter the Height of an object: m"))
Potential_Energy = Mass * (9.8) * Height
print(f"gravitational potential energy of an object: {Potential_Energy} joules")
```

```python
def calculate_potential_energy(mass, height):
    g = 9.8  # acceleration due to gravity in m/s²
    return mass * g * height

# Example usage
mass = 2
height = 10

energy = calculate_potential_energy(mass, height)
print(f"Potential Energy = {energy:.0f} J")
```

# Problem 4: Kinetic Energy of an Object

Problem Statement: Write a program to calculate the kinetic energy of an object in motion.

Formula:
Kinetic Energy = 1/2 m v²

where m is the mass and v is the velocity.

Input: Mass (in kilograms) and velocity (in meters per second).

Output: Kinetic energy (in joules).

Example:

Input: Mass = 10, Velocity = 3

Output: Kinetic Energy = 45 J

```python
Mass = int(input("Enter the Mass of an object: kg"))
Velocity = int(input("Enter the Velocity of an object: m/s"))
Kinetic_Energy = 1/2 * Mass * Velocity * Velocity
print(f"kinetic energy of an object: {Kinetic_Energy} J")
```

```python
def calculate_kinetic_energy(mass, velocity):
    return 0.5 * mass * velocity ** 2

# Example usage
mass = 10        # in kilograms
velocity = 3     # in meters per second

energy = calculate_kinetic_energy(mass, velocity)
print(f"Kinetic Energy = {energy:.0f} J")
```

# Problem 6: Projectile Motion – Maximum Height

Problem Statement: Write a program to calculate the maximum height reached by a
projectile, given its initial velocity and launch angle (ignoring air resistance).

Formula:
h_max = (v² × sin²(θ)) / (2g)

where v is the initial velocity, θ is the launch angle in degrees, and g = 9.8 m/s².

Input: Initial velocity (in meters per second) and angle (in degrees).

Output: Maximum height (in meters).

Example:

Input: Velocity = 20, Angle = 45

Output: Height = 20.4 meters

```python
# To make this program, we have to import math
import math
```

```python
velocity = float (input("Enter the value of velocity: m/s²"))
angle = float (input("Enter the value of angle: "))
g = 9.8 # Acceleration due to gravity (m/s²)

# for sin²0 , we use math.sin
# theta  is measured in degrees, but must be converted to radians
# Used math.radians(angle) to convert degrees to radians before applying math.sin.
angle_radians = math.radians(angle)

# Apply formula: h_max = (v² × sin²(θ)) / (2g)
h_max = (velocity**2) * (math.sin(angle_radians)**2) / (2*g)
print(f"Maximum height = {h_max:.2f} meters")
```

```python
import math

def calculate_max_height(velocity, angle_degrees):
    g = 9.8  # acceleration due to gravity in m/s²
    angle_radians = math.radians(angle_degrees)  # convert degrees to radians
    height = (velocity ** 2) * (math.sin(angle_radians) ** 2) / (2 * g)
    return height

# Example usage
velocity = 20       # in meters per second
angle = 45          # in degrees

max_height = calculate_max_height(velocity, angle)
print(f"Height = {max_height:.2f} meters")
```

# Problem 7: Harmonic Oscillator – Period of a Spring

Problem Statement: Write a program to calculate the period of a spring-mass system
(simple harmonic oscillator).

Formula:

T = 2π √(m/k)

where T is the period, m is the mass, and k is the spring constant.

Input: Mass (in kilograms) and spring constant k (in N/m).

Output: Period T (in seconds).

Example:

Input: Mass = 1, Spring constant = 50

Output: Period = 0.89 s

```python

# Input values
Mass = float(input("Enter the Mass (kg): "))
Spring_constant = float(input("Enter the spring constant (N/m): "))

# Calculate the period using the correct formula
Period = 2 * math.pi * math.sqrt(Mass / Spring_constant)

# Output the result
print(f"The period of a spring-mass system: {Period:.2f} s")
```

```python
import math

def calculate_period(mass, spring_constant):
    return 2 * math.pi * math.sqrt(mass / spring_constant)

# Example usage
mass = 1           # in kilograms
spring_constant = 50  # in N/m

period = calculate_period(mass, spring_constant)
print(f"Period = {period:.2f} s")
```

# Problem 8: Free Fall – Time to Hit the Ground

Problem Statement: Write a program to calculate the time it takes for an object to hit the
ground when dropped from a certain height (ignoring air resistance).

Formula:
t = √(2h/g)

where h is the height and g = 9.8 m/s².

Input: Height (in meters).

Output: Time to hit the ground (in seconds).

Example:

Input: Height = 20

Output: Time = 2.02 seconds

```python
height = h = float(input("Enter the height (in meters): "))
g = 9.8 # Constant for gravity
# Calculate time using the formula
t = math.sqrt(2 * h / g)

# Output the result rounded to 2 decimal places
print(f"time for an object: {t:.2f} seconds")
```

```python
import math

def calculate_fall_time(height):
    g = 9.8  # acceleration due to gravity in m/s²
    return math.sqrt((2 * height) / g)

# Example usage
height = 20  # in meters

time = calculate_fall_time(height)
print(f"Time to hit the ground = {time:.2f} seconds")
```

# Problem 9: Work Done by a Force

Problem Statement: Write a program to calculate the work done by a constant force acting
on an object.

Formula:
Work = F × d × cos(θ)

where F is the force, d is the displacement, and θ is the angle between the force and
displacement vectors.

Input: Force (in newtons), displacement (in meters), and angle (in degrees).

Output: Work done (in joules).

Example:

Input: Force = 100, Displacement = 10, Angle = 30

Output: Work = 866 J

```python
Force = F = float(input("Enter the Force: N"))
Displacement = D = float(input("Enter the Displacement: m"))
angle = float(input("Enter the angle (in degrees): "))

# Apply formula: Work = F × d × cos(θ)
# for cos(0) , we use math.cos
# theta  is measured in degrees, but must be converted to radians
# Used math.radians(angle) to convert degrees to radians before applying math.cos.
angle_radians = math.radians(angle)

# Apply formula: Work = F × d × cos(θ)
Work= F * D * (math.cos(angle_radians))
print(f"Work done: {Work:.0f} J")
```

```python
import math

def calculate_work(force, displacement, angle_degrees):
    angle_radians = math.radians(angle_degrees)  # convert degrees to radians
    return force * displacement * math.cos(angle_radians)

# Example usage
force = 100           # in newtons
displacement = 10     # in meters
angle = 30            # in degrees

work_done = calculate_work(force, displacement, angle)
print(f"Work = {work_done:.0f} J")
```

# Problem 10: Power Consumption of a Device

Problem Statement: Write a program to calculate the power consumed by an electrical
device.

Formula:
P = W / t

where P is power, W is work (energy), and t is time.

Input: Work (in joules) and time (in seconds).

Output: Power (in watts).

Example:

Input: Work = 500, Time = 50

Output: Power = 10 W

```python
# Input values
work = float(input("Enter the work done (in joules): "))
time = float(input("Enter the time taken (in seconds): "))

# Calculate power
power = work / time

# Output the result
print(f"Power = {power:.0f} W")
```

```python
def calculate_power(work, time):
    return work / time

# Example usage
work = 500     # in joules
time = 50      # in seconds

power = calculate_power(work, time)
print(f"Power = {power:.0f} W")
```
